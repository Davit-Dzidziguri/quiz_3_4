```python
import pandas as pd
import numpy as np
df = pd.read_csv('Earthquakes_database.csv')
df
```

```python
#1.წავშალეთ ყველა ის სვეტი, სადაც NaN მნიშვნელობა იყო მოცემული
df_empty =df.dropna(axis=1, how='any')
df_empty
```

```python
#2.დავყავით რიცხვითი მნიშვნელობის სვეტი კატეგორიებად
#გავუწერეთ ზღვრები
levels =[0, 50, 100, 500]
#დავყავით კატეგორიებად
categories=["Low", "Medium", "Large"]
#დავამატეთ ახალი სვეტი DepthDescription
df['DepthDescription']= pd.cut(df['Depth'], bins=levels, labels=categories,include_lowest=True)
df
```

```python
#3.
df.set_index('Location Source', inplace=True)
print(df)
```

```python
#გამოვიყენეთ stack() ფუნქცია location Source სვეტზე
stacked=df.stack()
print(stacked)
```

```python
#გამოვიყენეთ explode() ფუნქცია Longitude სვეტზე
explode = df.explode('Longitude')
print(explode)
```

```python
#4.Pie Chart
import matplotlib.pyplot as plt
first_seven = df.sort_values(by='Longitude', ascending=False).head(7)

plt.figure(figsize=(5, 5))
plt.pie(first_seven['Longitude'], labels=first_seven['Magnitude'])
plt.show()
```

```python
#Line Chart
df_ten = df.sample(n=10, random_state=50)

plt.figure(figsize=(10, 5))
plt.plot(df_ten['Date'], df_ten['Depth'],label='Depth')
plt.plot(df_ten['Date'],df_ten['Longitude'],label='Longitude')
plt.legend()
plt.show()
```

```python
#Scatter Chart
df_six = df.tail(6)
plt.figure(figsize=(8, 6))
plt.scatter(df_six['Magnitude'],df_six['Longitude'], color='green', marker='x',label='Magnitude')
plt.xlabel('Magnitude')
plt.ylabel('Longitude')
plt.legend()
plt.show()
```

```python
#Bar Chart
df_five = df.tail(5)
plt.figure(figsize=(5, 5))
plt.bar(df_five['Date'], df_five['Longitude'], color='brown',label='Longitude')
plt.bar(df_five['Date'], df_five['Depth'], color='cyan', label='Depth')
plt.legend()
plt.show()
```
